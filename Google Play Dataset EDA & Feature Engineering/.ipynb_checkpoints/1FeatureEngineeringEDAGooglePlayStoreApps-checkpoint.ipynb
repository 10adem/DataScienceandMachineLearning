```python
# Karakter temizlemek için oluşturulan kod bloğu
```

```python
chars_to_remove = ["+",",","$"]
cols_to_clean = ["Installs", "Price"]
```

```python
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_clean[cols] = df_clean[cols].str.replace(item, "")
```

```python
df_app_category = df_app_category.sort_values("Installs", ascending=False)
```

```python
apps = ["GAME", "COMMUNICATION", "TOOLS", "PRODUCTIVITY", "SOCIAL"]

plt.figure(figsize=(40,30))
for i, app in enumerate(apps):
    df_2 = df_app_category[df_app_category.Category == app]
    df_2 = df_2.head(5)
    #print(df_2)

    plt.subplot(3,2, i+1)
    sns.barplot(data=df_2, x="Installs", y="App")
    plt.title(app, size=30)
plt.tight_layout()
plt.show()
```
